```python
#selecting x and y

x = df['message']
y = df['Class']
```

```python
#importing library

from sklearn.model_selection import train_test_split
```

```python
#train test split with test size = 30%

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#importing Counvectorization from sklearn

from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Instantiate CountVectorizer with stop_words -vect is an object of class CountVectorizer()

vect = CountVectorizer(stop_words='english')
```

```python
# fit vect on the feature (X_train) -  converts a corpus of documents into a vector of unique words

x=vect.fit(x_train)
x
```

```python
#checkign the count of words in feature

x.vocabulary_
```

```python
#features named them separately as words

vect.get_feature_names()
```

```python
#transformming feature - train & test set

x_train_transform = vect.transform(x_train)
x_test_transform = vect.transform(x_test)
```

```python
#x_train after transformation

x_train_transform.toarray()
```

```python
#x_test after transformation

x_test_transform.toarray()
```

```python
#shape of final train set

x_train_transform.shape
```

```python
#printing  our final train set

print(x_train_transform)
```

```python
#printing feature names

print(vect.get_feature_names())
```

```python
#converting our x_train to dataframe with columns as vector's feature names

pd.DataFrame(x_train_transform.toarray(),columns=vect.get_feature_names())
```

```python
#import BernoulliNB

from sklearn.naive_bayes import BernoulliNB
```

```python
# instantiate bernoulli NB object

bnb = BernoulliNB()
```

```python
# fit model on training dataset

bnb.fit(x_train_transform,y_train)
```

```python
#model prediction

y_pred = bnb.predict(x_test_transform)
```

```python
#model prediction probability

y_pred_proba = bnb.predict_proba(x_test_transform)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#accuarcy score

metrics.accuracy_score(y_test,y_pred)
```

```python
#confusion metrics

confusion = metrics.confusion_matrix(y_test,y_pred)
confusion
```

```python
#Get True negative, False positive, False negative and True positive using confusion metrics

TN = confusion[0,0]
FP = confusion[0,1]
FN = confusion[1,0]
TP = confusion[1,1]
```

```python
# sensitivity using confusion metrics

sensitivity = TP/float(TP+FN)
sensitivity
```

```python
#specificity using confusion metrics

specificity = TN/float(TN+FP)
specificity
```

```python
#precision using confusion metrics

precision = TP/float(TP+FP)
precision
```

```python
#classification report

print(classification_report(y_test,y_pred))
```

```python
from flask import Flask,render_template,url_for,request
import pickle
import joblib


filename = 'model.pkl'
clf = pickle.load(open(filename, 'rb'))
cv=pickle.load(open('transform.pkl','rb'))
app = Flask(__name__)

@app.route('/')
def home():
	return render_template('home.html')

@app.route('/predict',methods=['POST'])
def predict():
	if request.method == 'POST':
		message = request.form['message']
		data = [message]
		vect = cv.transform(data).toarray()
		my_prediction = clf.predict(vect)
	return render_template('result.html',prediction = my_prediction)

if __name__ == '__main__':
	app.run()
```
